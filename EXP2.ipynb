```python
"""
Experiment Title

Data Cleaning and Feature Encoding using Pandas and Scikit-learn

Aim:

To perform data preprocessing on a raw dataset by applying data cleaning techniques and feature encoding methods using Python libraries such as Pandas and Scikit-learn, in order to prepare the dataset for machine learning model development.

Description:

In this experiment, a raw dataset is loaded and preprocessed through systematic data cleaning and feature transformation steps. The primary objective is to convert unstructured and inconsistent data into a clean, structured, and machine-learning-ready format.

The experiment includes the following stages:

Data Cleaning
Removal of duplicate records to avoid redundancy.
Handling missing values using appropriate imputation techniques:
Numerical columns are filled using the mean value.
Categorical columns are filled using the most frequent value.
Feature Encoding
Categorical variables are converted into numerical format using encoding techniques such as:
Label Encoding for ordinal or simple categorical features.
One-Hot Encoding (optional) for non-ordinal categorical variables.

Finally, the cleaned and encoded dataset is saved for further use in machine learning tasks such as classification or regression.

This experiment demonstrates the importance of preprocessing in data science workflows and highlights how proper data preparation improves model accuracy and reliability.

"""

# Import libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer

# Load dataset (change file name as needed)
df = pd.read_csv("raw_merged_heart_dataset.csv")

# Display first 5 rows
print("Original Data:")
print(df.head())

# -----------------------------
# 🔹 1. DATA CLEANING
# -----------------------------

# Remove duplicate rows
df = df.drop_duplicates()

# Handle missing values

# For numerical columns → fill with mean
num_imputer = SimpleImputer(strategy='mean')
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = num_imputer.fit_transform(df[num_cols])

# For categorical columns → fill with most frequent value
cat_imputer = SimpleImputer(strategy='most_frequent')
cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])

# -----------------------------
# 🔹 2. FEATURE ENCODING
# -----------------------------

# Label Encoding (for ordinal or simple categories)
label_encoder = LabelEncoder()

for col in cat_cols:
    df[col] = label_encoder.fit_transform(df[col])

# OR use One-Hot Encoding (better for non-ordinal data)
# df = pd.get_dummies(df, columns=cat_cols)

# -----------------------------
# 🔹 FINAL OUTPUT
# -----------------------------

print("\nCleaned & Encoded Data:")
print(df.head())

# Save cleaned dataset
df.to_csv("cleaned_data.csv", index=False)
```
